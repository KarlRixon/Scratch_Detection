```python
from PIL import Image
from matplotlib import pyplot as plt
from torchvision import transforms
```

```python
img_path = './train_data/train_img_{:04d}.jpg'
img_PIL
for i in range(1):
    img_PIL = Image.open(img_path.format(i+1))
    print('\npath: ', img_path.format(i+1), '\nsize: ', img_PIL.size, '\ntype: ',
          type(img_PIL), '\nformat: ', img_PIL.format, '\nmode: ', img_PIL.mode)
```

```python
def img_show(image, title=None):
    plt.figure()
    plt.imshow(image)
    if title is not None:
        plt.title(title)
        
img_show(img_PIL)
```

```python
# 裁剪到特定区域
defect1 = transforms.functional.crop(img_PIL, 437,4,550-437,177-4)
img_show(defect1, title='after_Crop')

defect2 = transforms.functional.crop(img_PIL, 366,400,402-366,645-400)
img_show(defect2, title='after_Crop')

defect3 = transforms.functional.crop(img_PIL, 410,344,585-410,518-344)
img_show(defect3, title='after_Crop')
```

```python
toTensor = transforms.Compose([transforms.ToTensor()])

# 尺寸变化、缩放
transform_scale = transforms.Compose([transforms.Resize(64)])
temp = transform_scale(defect2)
plt.figure()
img_show(temp, title='after_scale')
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import os
from torchvision import transforms
import random

f_out = open('tail_train.txt', 'w')
f_out2 = open('tail_test.txt', 'w')
f_in = open('train_labels.txt', 'r')
count_defect = 0
count_intact = 0
count_all = 0
h_avg = 0
w_avg = 0
for line in f_in:
    
    count_all += 1
    
    if not os.path.exists('tail_data/'):
        os.makedirs('tail_data/')
        
    toTensor = transforms.Compose([transforms.ToTensor()])
    transform_scale = transforms.Compose([transforms.Resize(64)])
    
    if 'NG' in line:
        if not os.path.exists('tail_data/defect/'):
            os.makedirs('tail_data/defect/')
        
        # 获取图片路径
        path = line[-31:]
        
        # 获取裁剪坐标
        coordinates = line[19:-33]
        for obj in coordinates.split('],['):
            ymin = int(obj.split(',')[1])
            xmin = int(obj.split(',')[0])
            ymax = int(obj.split(',')[3])
            xmax = int(obj.split(',')[2])
            
            # 裁剪并保存并填写信息
            img = Image.open(path[:-1])
            # defect = transforms.functional.crop(img, ymin, xmin, ymax-ymin, xmax-xmin)  # i, j, h, w
            # 处理为正方形便于缩放，划痕居中
            max_hw = max(ymax-ymin, xmax-xmin)
            if ymax-ymin >= xmax-xmin:
                new_x = xmin - int((ymax-ymin-xmax+xmin)/2)
                defect = transforms.functional.crop(img, ymin, new_x, ymax-ymin, ymax-ymin)  # i, j, h, w
            if ymax-ymin < xmax-xmin:
                new_y = ymin - int((xmax-xmin-ymax+ymin)/2)
                defect = transforms.functional.crop(img, new_y, xmin, xmax-xmin, xmax-xmin)  # i, j, h, w
            
            count_defect += 1
            defect = transform_scale(defect)
            defect.save('tail_data/defect/'+str(count_defect)+'.jpg')
            # 区分训练集和测试集
            if random.randint(1,100) > 15:
                f_out.write('tail_data/defect/'+str(count_defect)+'.jpg '+'1\n')
            else:
                f_out2.write('tail_data/defect/'+str(count_defect)+'.jpg '+'1\n')
            
            # 统计标记区域的平均大小
            h_avg += ymax-ymin
            w_avg += xmax-xmin

            # 查看裁剪图的原图
#             if count_defect == 476:
#                 print(path)

    elif 'OK' in line:
        if not os.path.exists('tail_data/intact/'):
            os.makedirs('tail_data/intact/')
        
        # 获取图片路径
        path = line[-31:]
        
        # 获取n个随机裁剪坐标 h=w=100
        for i in range(6):
            x = random.randint(0, 712)
            y = random.randint(0, 517)
        
            # 裁剪并保存并填写信息
            img = Image.open(path[:-1])
            defect = transforms.functional.crop(img, y, x, 100, 100)
            defect = transform_scale(defect)
            
            count_intact += 1
            defect.save('tail_data/intact/'+str(count_intact)+'.jpg')
            # 区分训练集和测试集
            if random.randint(1,100) > 15:
                f_out.write('tail_data/intact/'+str(count_intact)+'.jpg '+'0\n')
            else:
                f_out2.write('tail_data/intact/'+str(count_intact)+'.jpg '+'0\n')
            
    else:
        print('train_labels.txt error at line ', count_all)
        break
        
f_out.close()
f_out2.close()
f_in.close()
print('average hight: ', h_avg/count_defect) # 102.90787589498807
print('average width: ', w_avg/count_defect) # 85.04367541766109
print('defect: ', count_defect)
print('intact: ', count_intact)
print('all: ', count_all, ' -> ', count_defect+count_intact)
```
